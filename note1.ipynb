```python
def fillMissing(df):
    temp = df.copy()
    numeric_cols = temp.select_dtypes(include='number').columns
    temp[numeric_cols] = temp[numeric_cols].fillna(temp[numeric_cols].mean())
    return temp
```

```python
df = fillMissing(df)
```

```python
df = df.drop_duplicates()
```
